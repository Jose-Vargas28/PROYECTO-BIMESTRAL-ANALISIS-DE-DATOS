```python
import json
import pymongo

with open("RestDB.json", "r", encoding="utf-8") as file:
    data = json.load(file) 
mongo_client = pymongo.MongoClient("mongodb://localhost:27017/")
mongo_db = mongo_client["RestaurantDB"]  
mongo_collection = mongo_db["RestCollection"]  
if isinstance(data, list):  
    mongo_collection.insert_many(data)
else:  
    mongo_collection.insert_one(data)

print("Migración completada")

mongo_client.close()

```
