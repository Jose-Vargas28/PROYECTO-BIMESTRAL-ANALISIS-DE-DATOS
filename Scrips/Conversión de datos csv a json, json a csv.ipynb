```python
import requests
import json
from requests.auth import HTTPBasicAuth

# URL de la base de datos en CouchDB
db_url = "http://localhost:5984/migracion/_bulk_docs"

# Credenciales de acceso a CouchDB
username = 'admin'
password = '123456'

# Leer el archivo línea por línea y convertirlo a una lista de documentos
docs = []
with open('Hobby_Data.json', 'r') as f:
    for line in f:
        docs.append(json.loads(line.strip()))  # Cada línea es un documento JSON

# Empaquetar los documentos en un objeto JSON con la clave "docs"
data = {"docs": docs}

# Enviar la solicitud POST para insertar los documentos con autenticación básica
response = requests.post(db_url, json=data, auth=HTTPBasicAuth(username, password))

# Ver la respuesta
if response.status_code == 200:
    print("¡Documentos importados exitosamente!")
else:
    print(f"Error al importar documentos: {response.text}")
```

```python
import requests
import json
from requests.auth import HTTPBasicAuth

# URL de la base de datos en CouchDB
db_url = "http://localhost:5984/migracion/_all_docs?include_docs=true"

# Credenciales de acceso a CouchDB
username = 'admin'
password = '123456'

# Realizar la solicitud GET para obtener todos los documentos
response = requests.get(db_url, auth=HTTPBasicAuth(username, password))

# Verificar la respuesta
if response.status_code == 200:
    # Obtener los documentos del cuerpo de la respuesta
    docs = response.json()['rows']
    
    # Extraer los documentos completos (incluyendo los datos bajo la clave 'doc')
    documents = [doc['doc'] for doc in docs]

    # Guardar los documentos en un archivo JSON
    with open('datos_exportados.json', 'w') as f:
        json.dump(documents, f, indent=4)
    
    print("¡Documentos exportados exitosamente!")
else:
    print(f"Error al exportar documentos: {response.text}")
```

```python
import requests
import csv
from requests.auth import HTTPBasicAuth

# URL de la base de datos en CouchDB
db_url = "http://localhost:5984/migracion/_all_docs?include_docs=true"

# Credenciales de acceso a CouchDB
username = 'admin'
password = '123456'

# Realizar la solicitud GET para obtener todos los documentos
response = requests.get(db_url, auth=HTTPBasicAuth(username, password))

# Verificar la respuesta
if response.status_code == 200:
    # Obtener los documentos del cuerpo de la respuesta
    docs = response.json()['rows']
    
    # Extraer los documentos completos (incluyendo los datos bajo la clave 'doc')
    documents = [doc['doc'] for doc in docs]
    
    # Abrir el archivo CSV para escribir los datos
    with open('datos_exportados_hobbies.csv', 'w', newline='') as csvfile:
        # Crear un escritor CSV
        fieldnames = documents[0].keys()  # Usar las claves del primer documento como las columnas
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        # Escribir el encabezado (nombres de las columnas)
        writer.writeheader()

        # Escribir los documentos como filas en el archivo CSV
        for document in documents:
            writer.writerow(document)

    print("¡Documentos exportados exitosamente como CSV!")
else:
    print(f"Error al exportar documentos: {response.text}")
```
